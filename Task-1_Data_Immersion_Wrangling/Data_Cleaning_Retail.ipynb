```python
from multiprocessing.reduction import duplicate

# Import Required Library #
import pandas as pd
```

```python
# Load Dataset #
df=pd.read_excel("Superstore_Sales.xls", engine="xlrd")
print("Dataset Loaded Successfully")
print("-"*50)
```

```python
print("\nDataset Info:")
display(df.info())
```

```python
# Convert Postal Code to String #
df['Postal Code']=df['Postal Code'].astype(str)
```

```python
# Verify #
df.dtypes
```

```python
# Statistical Summary #
print("\nStatistical Summary (Numerical columns):")
display(df.describe())
```

```python
print("\nStatistical Summary (Including Categorical columns):")
display(df.describe(include='all'))
```

```python
# Check Missing Values #
print("\nMissing values:")
display(df.isnull().sum())
```

```python
# Remove Duplicates #
duplicate=df.duplicated().sum()
print(f"\nNumber of rows:{duplicate}")
```

```python
# Cleaning & Categorizing Free-Text Fields #
df['Segment']=df['Segment'].str.strip().str.title()
df['Category']=df['Category'].str.strip().str.title()
df['Sub-Category']=df['Sub-Category'].str.strip().str.title()
df['Region']=df['Region'].str.strip().str.title()

```

```python
# Create Region Group (Categorization) #

def Region_group(Region):
    if Region in ['East', "West"]:
        return "Coastral"
    else:
        return "Central"
df['Region']=df['Region'].apply(Region_group)

df.columns


```

```python
# Feature Engineering #
df['Order Year']=df['Order Date'].dt.year
df['Order Month']=df['Order Date'].dt.month
df.columns
```

```python
# Save cleaned Dataset #
df.to_csv("Cleaned_Dataset.csv", index=False)
print("\nDataset Cleaned Successfully")
print("-"*50)
```
